# 00 Preparation
Laden der f√ºr die Bearbeitung notwendigen Bibliotheken üìö.

```python
# IMPORT LIBRARIES
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

# 01 Load Data
Laden des Dataset üí≥:

```python
# LOAD DATA
df = pd.read_csv('09_disaster_smote.csv')

# ENTFERNEN DER SPALTE DATE
df.drop(['Date'], axis=1, inplace=True)

# DATE UMWANDELN IN DATETIME
#df['Date'] = pd.to_datetime(df['Date'])
```

```python
# VERTEILUNG DER WERTE IN ZIELVARIABLE
df['DisasterOccurred'].value_counts()
```

# 02 SMOTE Anwenden
Mittels <code>SMOTE</code> Synthetic Minority Over-sampling Technique die Anzahl an Katastrophen in dem Dataset erh√∂hen, um es ausgeglichener ‚öñÔ∏è zu machen.

```python
# TRENNEN FEATURES UND ZIELVARIABLE
X = df.drop('DisasterOccurred', axis=1)
Y = df['DisasterOccurred']
```

```python
# NEUES SMOTE OBJEKT ERSTELLEN, UM ANZAHL AUSZUGLEICHEN
smote = SMOTE(sampling_strategy=0.5, random_state=1234)

# SMOTE ANWENDEN
X_resampled, Y_resampled = smote.fit_resample(X, Y)
```

```python
# SYNTHETISCHE FEATURES UND SYNTHEISCHE ZIELVARIABLE WIEDER ZUSAMMENF√úGEN
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(Y_resampled, columns=['DisasterOccurred'])], axis=1)

# TRENNEN FEATURES UND ZIELVARIABLE MIT SMOTE
X_smote = df_resampled.drop('DisasterOccurred', axis=1)
Y_smote = df_resampled['DisasterOccurred']
```

```python
# VERTEILUNG DER WERTE IN ZIELVARIABLE
df_resampled['DisasterOccurred'].value_counts()
```

# 03 Split Data
Aufteilen der Daten ‚úÇÔ∏è in ein Trainings- und Testdatenset. Einmal mit und einmal ohne SMOTE.

```python
# SPLIT in 70% TRAIN und 30% TEST OHNE SMOTE
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=1234, stratify=df['DisasterOccurred'])
```

```python
# SPLIT in 70% TRAIN und 30% TEST MIT SMOTE
X_train_smote, X_test_smote, Y_train_smote, Y_test_smote = train_test_split(
    X_smote, Y_smote, test_size=0.3, random_state=1234, stratify=df_resampled['DisasterOccurred'])
```

# 04 Train Model
Die Modelle jeweils mit einem **Decision Tree** üå≤ trainieren üí™.

```python
# DECISION TREE CLASSIFIER OHNE SMOTE
dtc = DecisionTreeClassifier(random_state=1234)
dtc.fit(X_train, Y_train)
```

```python
# DECISION TREE CLASSIFIER MIT SMOTE
dtc = DecisionTreeClassifier(random_state=1234)
dtc.fit(X_train_smote, Y_train_smote)
```

```python
# VORHERSAGEN MACHEN OHNE SMOTE
Y_pred = dtc.predict(X_test)
```

```python
# VORHERSAGEN MACHEN MIT SMOTE
Y_pred_smote = dtc.predict(X_test_smote)
```

# 05 Modell Evaluieren
Die Kennzahlen **Accuracy** und **Recall berechnen**, um zu bestimmen, wie gut unsere Vorhersagen sind üßê.

```python
# KENNZAHLEN BERECHNEN OHNE SMOTE
accuracy = accuracy_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)

# KENNZAHLEN BERECHNEN MIT SMOTE
accuracy_smote = accuracy_score(Y_test_smote, Y_pred_smote)
recall_smote = recall_score(Y_test_smote, Y_pred_smote)
precision_smote = precision_score(Y_test_smote, Y_pred_smote)

# KENNZAHLEN AUSGEBEN
print(f"Accuracy: {accuracy:.2f}\nRecall: {recall:.2f}\nPrecision: {precision:.2f}\n")
print(f"Accuracy SMOTE: {accuracy_smote:.2f}\nRecall SMOTE: {recall_smote:.2f}\nPrecision SMOTE: {precision_smote:.2f}")
```
