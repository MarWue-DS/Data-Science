### 1. üì• Import
Lade folgende Daten <code>04_kunden_daten.csv</code> und pr√ºfe auf fehlende Werte.

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.model_selection import train_test_split

# LOAD DATA
df = pd.read_csv("04_kunden_daten.csv")
df
```

### 2. üîé Fehlende Werte finden
Wie viele fehlen in jeder Spalte? Ersetze diese mit dem Median und Mode.

```python
# MISSING VALUES
print(df.isnull().sum())
```

```python
# REPLACE MEDIAN
df_ek = df.fillna(df.mean(numeric_only=True))
mode_bildung = df['Bildung'].mode()[0]
df_bi = df_ek.copy()
df_bi['Bildung'] = df_bi['Bildung'].fillna(mode_bildung)
df_bi
```

### 3. üî• One-Hot-Encoding
Wandle kategorische Spalten in numerische Spalten um.

```python
# ONE-HOT-ENCODING
df_encoded = pd.get_dummies(df_bi, columns=['Bildung'], dtype=int)
df_encoded
```

### ‚úÇÔ∏è 4. Split 
Teile die Daten in 80% Training und 20% Testdaten auf.

```python
# FEATRUES UND TARGET
X = df_encoded.drop(['Kaufkraft'], axis=1)
Y = df_encoded['Kaufkraft']

# SPLIT DER DATEN
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1234)
```

### üí™ 5. Training
Trainiere ein <code>LinearRegression</code> und ein <code>KNeighborsRegressor</code> Modell.

```python
# TRAIN MODEL LINEAR REGRESSION
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, Y_train)


# TRAIN MODEL KNEIGHBORS REGRESSOR
from sklearn.neighbors import KNeighborsRegressor

nr = KNeighborsRegressor()
nr.fit(X_train, Y_train)


# PREDICTIONS
Y_pred_lr = lr.predict(X_test)
Y_pred_nr = nr.predict(X_test)
```

### üßÆ 6. Kennzahlen 
Berechne f√ºr beide Modelle: MSE, RMSE, MAE, R¬≤.

```python
# CALCLUATE METRICS LR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

mse_lr = mean_squared_error(Y_test, Y_pred_lr)
rmse_lr = np.sqrt(mse_lr)
mae_lr = mean_absolute_error(Y_test, Y_pred_lr)
r2_lr = r2_score (Y_test, Y_pred_lr)

# CALCLUATE METRICS NR
mse_nr = mean_squared_error(Y_test, Y_pred_nr)
rmse_nr = np.sqrt(mse_nr)
mae_nr = mean_absolute_error(Y_test, Y_pred_nr)
r2_nr = r2_score (Y_test, Y_pred_nr)
```

```python
# OUTPUT PREDICTIONS
comparison_df = pd.DataFrame ({
    'Original' : Y_test,
    'Linear Regression' : Y_pred_lr,
    'KNeighbors Regressor' : Y_pred_nr
}, index=X_test.index)
comparison_df
```

```python
# OUTPUT METRICS
print(f"MSE LR: {mse_lr:.2f}")
print(f"MSE NR: {mse_nr:.2f}")

print(f"RMSE LR: {rmse_lr:.2f}")
print(f"RMSE NR: {rmse_nr:.2f}")

print(f"MAE LR: {mae_lr:.2f}")
print(f"MAE NR: {mae_nr:.2f}")

print(f"R¬≤ LR: {r2_lr:.2f}")
print(f"R¬≤ NR: {r2_nr:.2f}")
```

### üåü BONUS
Was bedeutet ein negativer R¬≤ Wert?

Ein negativer ( R^2 )-Wert ist ein Hinweis darauf, dass das Modell nicht gut funktioniert und √ºberarbeitet werden muss. Es bietet eine Gelegenheit, die Leistung des Modells zu analysieren und m√∂gliche Verbesserungen zu identifizieren.
