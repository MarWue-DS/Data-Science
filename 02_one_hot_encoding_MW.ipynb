# üî• One-Hot-Encoding
### Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd

# CREATE DATA
df = pd.DataFrame({
    'Farbe': ['rot', 'blau', 'gr√ºn', 'gelb']
})
df
```

### üêº One-Hot-Encoding mit Pandas

```python
# ONE-HOT-ENCODINGG
encoded = pd.get_dummies(df, columns=['Farbe'], dtype=int)
encoded
```

### üéì One-Hot-Encoding mit lediglich 2 Werten
Der Parameter <code>drop_first=True</code> bei **One-Hot-Encoding** entfernt die erste Kategorie jeder Spalte, um Redundanz zu vermeiden und Multikollinearit√§t in linearen Modellen zu reduzieren.

```python
# CREATE DATA
# CREATE DATA
df = pd.DataFrame({
    'Sex': ['weiblich', 'm√§nnlich']
})

# ONE-HOT-ENCODING
encoded = pd.get_dummies(df, columns=['Sex'], dtype=int, drop_first=True)
encoded
```

### ‚úÖ Aufgabe
- Erstelle ein DataFrame mit einer Spalte Tier (["Hund", "Katze", "Maus", "Vogel"])
- Wandle die Spalte per <code>pd.get_dummies()</code> in One-Hot-Encoding um

```python
# CREATE DATA
df = pd.DataFrame ({
    'Tier' : ["Hund", "Katze", "Maus", "Vogel"]
})

# ONE-HOT-ENCODINGG
encoded = pd.get_dummies(df, columns=['Tier'], dtype=int,)
encoded
```
