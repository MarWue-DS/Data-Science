```python
# IMPORT LIBRARIES
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# LAOD DATA
df = pd.read_csv('05_titanic_data.csv')
df
```

```python
# VISUALIZE
sns.boxplot(x=df['Age'])
plt.title('Alter Verteilung VORHER')
plt.show()
```

```python
# OUTLIER
cap_value = 64
threshold = 64

# LAMBDA-FUNCTION
df['Age'] = df['Age'].apply(lambda x: cap_value if x > threshold else x)
```

```python
# VISUALIZE
sns.boxplot(x=df['Age'])
plt.title('Alter Verteilung NACHHER')
plt.show()
```

```python
# PLOT GRAPH
plt.hist(x=df['Age'], edgecolor='black', bins=6)
plt.grid(True, alpha=.25)
```
