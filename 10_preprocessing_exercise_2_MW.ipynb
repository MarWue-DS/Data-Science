# 02 Fehlende Werte
Existieren **fehlende Werte** in dem Dataset? Falls ja und √ºbersteigen diese die Anzahl von 10, entferne üóë diese Spalten.

```python
# ALLE SPALTEN MIT FEHLENDEN WERTEN FINDEN
missing = df.isnull().sum()
missing
```

```python
# KOPIE ANFERTIGEN
df_delete = df.copy()

# ZEILEN MIT FEHLENDEN WERTEN L√ñSCHEN
df_delete.dropna(thresh=len(df_delete) - 10, axis=1)

# MODE F√úR RESTLICHE SPALTEN
for column in df_delete.select_dtypes(include=['object']).columns:
    if df_delete[column].isnull().sum() > 0:  # Pr√ºfen, ob fehlende Werte vorhanden sind
        modus_wert = df_delete[column].mode()[0]
        df_delete[column] = df_delete[column].fillna(modus_wert)

df_delete = df_delete.infer_objects()

# MEAN F√úR RESTLICHE SPALTEN
for column in df_delete.select_dtypes(include=['number']).columns:
    if df_delete[column].isnull().sum() > 0:  # √úberpr√ºfen, ob fehlende Werte vorhanden sind
        mean_wert = df_delete[column].mean()
        df_delete[column] = df_delete[column].fillna(mean_wert)

print("√úberpr√ºfen auf fehlende Werte:")
print(df_delete.isnull().sum())
```
