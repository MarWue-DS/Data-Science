# 00 Preparation
Laden der f√ºr die Bearbeitung notwendigen Bibliotheken üìö.

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import numpy as np
```

# 01 Load Data
Laden des Dataset üí≥:

```python
# LOAD DATA
df = pd.read_csv('05_creditcard_defaults.csv')

# SPALTE ID ENTFERNEN (vorher Sicherungskopie machen)
df_cleaned = df.copy()
df_cleaned.drop(['ID'], axis=1, inplace=True)

df_cleaned['Pay Stat M1'].unique()
```

```python
df_cleaned
```

# 02 Fehlende Werte
Existieren **fehlende Werte** in dem Dataset? Falls ja, entferne üóë diese Zeilen.

```python
# ALLE ZEILEN MIT FEHLENDEN WERTEN L√ñSCHEN
df_cleaned = df_cleaned.dropna()
```

# 03 Normalize Data
**Normalisieren** alle numerischen Werte mit Hilfe von <code>MinMax</code>.

```python
# INITIALIZE MINMAX SCLAER (vorher Sicherungskopie machen)
df_normal = df_cleaned.copy()
scaler = MinMaxScaler()

# NUMERISCHE SPALTEN SELEKTIEREN
numeric_columns = df_normal.select_dtypes(include=['float64', 'int64']).columns

# SELEKTIERTE SPALTEN NORMALISIEREN
df_normal[numeric_columns] = scaler.fit_transform(df_normal[numeric_columns])
```

# 04 Split Data
Aufteilen der Daten ‚úÇÔ∏è in ein Trainings- und Testdatenset.

```python
# AUFZEILEN IN FEATURES UND ZIELVARIABLE
X = df_normal.drop(['Default Next Month'], axis=1)
Y = df_normal['Default Next Month']
Y = df_normal['Default Next Month'].map({'No': 0, 'Yes': 1})
Y = np.ravel(Y)

# SPLIT in 70% TRAIN und 30% TEST
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234, stratify=df_normal['Default Next Month'])
```

# 05 Train Model
Das Model mit einer **Logistischen Regression** trainieren üí™.

```python
# LOGISTISCHE REGRESSION
lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
# VORHERSAGEN MACHEN
Y_pred = lr.predict(X_test)
Y_test
```

# 06 Modell Evaluieren
Die Kennzahlen **Accuracy** und **Recall berechnen**, um zu bestimmen, wie gut unsere Vorhersagen sind üßê.

```python
# KENNZAHLEN BERECHNEN
accuracy = accuracy_score(Y_test, Y_pred)
recall = recall_score(Y_test, Y_pred)
precision = precision_score(Y_test, Y_pred)
f1 = f1_score(Y_test, Y_pred)


# KENNZAHLEN AUSGEBEN
print(f"Accuracy: {accuracy:.2f}\nRecall: {recall:.2f}\nPrecision: {precision:.2f}\nF1-Score: {f1:.2f}")
```

```python
# IMPORT AND TRAIN DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state=1234)

# IMPORT AND TRAIN RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=1234)

# IMPORT AND TRAIN SUPPORT VECTOR CLASSIFIER
from sklearn.svm import SVC
svc = SVC(kernel="rbf", gamma=0.5)

# IMPORT AND TRAIN LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=12345, max_iter=10000)
```

```python
# IMPORT GRID SEARCH
from sklearn.model_selection import GridSearchCV
```

```python
# GRID F√úR DECISION TREE
dtc_param = {
    "criterion": ["gini", "entropy"],
    "min_samples_split": [8, 16],
    "min_samples_leaf": [3, 4, 5, 6, 7]
}

dtc_grid = GridSearchCV(
    estimator=dtc,
    param_grid=dtc_param,
    scoring="recall",
    cv=10,
    return_train_score=True
)
```

```python
# GRID F√úR RANDOM FOREST
rfc_param = {
    "n_estimators": [10, 15, 20],
    "min_samples_split": [8, 16],
    "min_samples_leaf": [3, 4, 5, 6, 7]
}

rfc_grid = GridSearchCV(
    estimator=rfc,
    param_grid=rfc_param,
    scoring="recall",
    cv=10,
    return_train_score=True
)
```

```python
# GRID F√úR LOGISTIC REGRESSION
lr_param = {
    "C": [0.01, 0.1, 0.5, 1, 2, 5, 10],
    "penalty": ["l2"],
    "solver": ["liblinear", "lbfgs", "saga"]
}

lr_grid = GridSearchCV(
    estimator=lr,
    param_grid=lr_param,
    scoring="recall",
    cv=10,
    return_train_score=True
)
```

```python
# # GRID F√úR SVC 
svc_param = {
    "C": [0.01, 0.1, 0.5, 1, 2, 5, 10],
    "kernel": ["rbf", "linear"],
    "gamma": [0.1, 0.25, 0.5, 1, 5]
}

svc_grid = GridSearchCV(
    estimator=svc,
    param_grid=svc_param,
    scoring="recall",
    cv=10,
    return_train_score=True
)
```

```python
# DECISION TREE TRAINIEREN MIT GRID SEARCH
dtc_grid_fit = dtc_grid.fit(X,Y)
```

```python
# RANDOM FOREST TRAINIEREN MIT GRID SEARCH
rfc_grid_fit = rfc_grid.fit(X,Y)
```

```python
# LOGISTIC REGRESSION TRAINIEREN MIT GRID SEARCH
lr_grid_fit = lr_grid.fit(X,Y)
```

```python
# SVC TRAINIEREN MIT GRID SEARCH
svc_grid_fit = svc_grid.fit(X,Y)
```

```python
# FUNKTION, UM RESULTATE ZU VERGLEICHEN
def best_grid_scores(grid_fit):
    cv_result = pd.DataFrame(grid_fit.cv_results_)

    # Nach bestem Test-Score und Zeit sortieren
    cv_result = cv_result.sort_values(by=['rank_test_score', 'mean_score_time'])
    best_result = cv_result.iloc[0][['mean_test_score', 'mean_train_score']]

    # Umbenennen und transponieren
    best_result = best_result.rename({
        'mean_test_score': 'Tuning Avg. Test Score',
        'mean_train_score': 'Tuning Avg. Train Score'
    }).to_frame().T

    return best_result.transpose()

# FUNKTION F√úR ALLE MODELLE AUFRUFEN
lr_result = best_grid_scores(lr_grid_fit)
dtc_result = best_grid_scores(dtc_grid_fit)
rfc_result = best_grid_scores(rfc_grid_fit)
svc_result = best_grid_scores(svc_grid_fit)

print(lr_result)
print(dtc_result)
print(rfc_result)
print(svc_result)
```
