```python
# IMPORT LIBRARIES
import pandas as pd
from imblearn.over_sampling import SMOTE
```

# 01 Load Data
Laden des Dataset üè¶.

```python
# LOAD DATA
df = pd.read_csv('02_loan_data.csv')
```

```python
# VERTEILUNG DER WERTE IN ZIELSPALTE VOR SMOTE
df['Loan_Status'].value_counts()
```

# 02 SMOTE Anwenden
Mittels <code>SMOTE</code> Synthetic Minority Over-sampling Technique die Anzahl an nicht-genehmigten Darlehen in dem Dataset erh√∂hen, um es ausgeglichener ‚öñÔ∏è zu machen.

```python
# TRENNEN FEATURES UND ZIELVARIABLE
X = df.drop('Loan_Status', axis=1)
Y = df['Loan_Status']
```

```python
# NEUES SMOTE OBJEKT ERSTELLEN, UM ANZAHL ABGELEHNTER UND GENEHMIGTER DARLEHEN AUSZUGLEICHEN
smote = SMOTE(sampling_strategy='auto', random_state=1234)

# SMOTE ANWENDEN
X_resampled, Y_resampled = smote.fit_resample(X, Y)
```

```python
# SYNTHETISCHE FEATURES UND SYNTHEISCHE ZIELVARIABLE WIEDER ZUSAMMENF√úGEN
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(Y_resampled, columns=['Loan_Status'])], axis=1)
df_resampled
```

# 03 √úberpr√ºfen der Verteilung
√úberpr√ºfen, ob sich das Verh√§tnis der Daten nach Anwendung von SMOTE ver√§ndert hat ü§î.

```python
# VERTEILUNG DER WERTE IN ZIELSPALTE NACH SMOTE
df_resampled['Loan_Status'].value_counts()
```

# 04 BONUS: Explizite Angabe 
Explizit Angabe ‚≠ï, wie viele Datens√§tze SMOTE von jeder Klasse jeweils erzeugen soll.

```python
# NEUES SMOTE OBJEKT ERSTELLEN, UM ANZAHL ABGELEHNTER UND GENEHMIGTER DARLEHEN AUSZUGLEICHEN
smote = SMOTE(sampling_strategy={'N': 100, 'Y': 250}, random_state=1234)


# SMOTE ANWENDEN
X_resampled, Y_resampled = smote.fit_resample(X, Y)

# SYNTHETISCHE FEATURES UND SYNTHEISCHE ZIELVARIABLE WIEDER ZUSAMMENF√úGEN
df_resampled = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(Y_resampled, columns=['Loan_Status'])], axis=1)


# VERTEILUNG DER WERTE IN ZIELSPALTE NACH SMOTE
df_resampled['Loan_Status'].value_counts()
```
