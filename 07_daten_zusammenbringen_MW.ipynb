# 01 Daten zusammenf√ºgen
In Pandas k√∂nnen wir DataFrames auf verschiedene Arten zusammenf√ºgen:
- **concat:** F√ºgt DataFrames aneinander ‚Äì entweder Zeile f√ºr Zeile oder Spalte f√ºr Spalte.
- **merge:** Kombiniert DataFrames basierend auf gemeinsamen Spalten, √§hnlich wie ein "inner join‚Äú in einer Datenbank. 

```python
# PANDAS IMPORTIEREN
import pandas as pd
```

### Daten laden
Laden der beiden Files.

```python
# DATAFRAMES LADEN
df_person = pd.read_csv('06_person_daten.csv')
df_adress = pd.read_csv('06_adress_daten.csv')

df_adress
```

### Merge
Zusammenf√ºgen der beiden Tabellen.

```python
# DATAFRAMES MERGEN
df_merged = pd.merge(df_person, df_adress, on='ID', how='inner')
df_merged
```

### Daten anh√§ngen
Neue Daten an bestehende Daten anh√§ngen.

```python
# DATAFRAME LADEN
df_person_neu = pd.read_csv("06_person_neu.csv")
df_person_neu
```

```python
# DATAFRAME ANH√ÑNGEN
df_person2 = pd.concat([df_person, df_person_neu], ignore_index=True)
df_person2
```
