# 01 Load Data

```python
# LOAD DATA
df_personal = pd.read_csv('10a_personaldaten.csv')
```

```python
# LOAD DATA
df_address = pd.read_csv('10b_adressdaten.csv')
print("Personal Data:\n", df_personal.head())
print("\nAddress Data:\n", df_address.head())
```

# 02 Merge Data
Beide DataFrames ğŸ”— **zusammenbringen** und das Ergebnis in einem neuen DataFrame speichern. Der Merge soll Ã¼ber die Spalte stattfinden, die in beiden DataFrames enthalten ist.

```python
# MERGE BEIDER DATAFRAMES
df_merged = pd.merge(df_personal, df_address, on='ID', how="left")

# ÃœberprÃ¼fe das Ergebnis des Merges
print("\nMerged Data:\n", df_merged.head())
```

# 03 Daten Transformatieren
Lege eine neue Spalte **Generation** ğŸ†• auf Basis des Alters (AGE).

```python
# NAMEN FÃœR GENERATION VERGEBEN


# GRENZEN FÃœR GENERATION DEFINIEREN (0-26: Gen Z, 27-42: Gen Y, 43-57: Gen X, >58: Boommer)
def determine_generation(age):
    if age < 26:
        return 'Gen Z'
    elif 26 <= age < 42:
        return 'Gen Y'
    elif 42 <= age < 57:
        return 'Gen X'
    else:
        return 'Boomer'

# Anwenden der Funktion auf die AGE-Spalte
df_merged['Generation'] = df_merged['AGE'].apply(determine_generation)

# ÃœberprÃ¼fen des DataFrames nach HinzufÃ¼gen der Generation
print("\nDataFrame mit Generation:\n", df_merged.head())

# NEUE SPALTE ANLEGEN

```

```python
# Durchschnittsalter fÃ¼r jede Generation berechnen
average_age_per_gen = df_merged.groupby('Generation')['AGE'].mean().reset_index()
average_age_per_gen.rename(columns={'AGE': 'AVG_AGE'}, inplace=True)

# Merge mit dem ursprÃ¼nglichen DataFrame, um die AVG_AGE-Spalte hinzuzufÃ¼gen
df_merged2 = df_merged.merge(average_age_per_gen, on='Generation', how='left')

# Ausgabe des aktualisierten DataFrames
print("\nDataFrame mit AVG_AGE:\n", df_merged2.head())
```
