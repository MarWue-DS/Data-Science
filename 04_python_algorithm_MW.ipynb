# üí™ Train Modell Data
### Insurance Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.model_selection import train_test_split

# LOAD DATA
df = pd.read_csv("03_insurance_data.csv")
df
```

### üêº One-Hot-Encoding mit Pandas

```python
# ONE-HOT-ENCODING
df_encoded = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], dtype=int)
df_encoded
```

```python
# FEATURES & TARGET
X = df_encoded.drop(['charges'], axis=1)
Y = df_encoded['charges']
```

### ‚úÇÔ∏è Insurance Daten aufteilen
Daten splitten in 70% Training und 30% Testdaten.

```python
# DATEN IN TRAIN UND TEST SPLITTEN
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1234)
```

### üèãüèª‚Äç‚ôÇÔ∏è Modell trainieren
Das ML-Modell mit einem Algorithmus deiner Wahl trainieren, um Vorhersagen treffen zu k√∂nnen.

```python
# IMPORT LIBRARIES
from sklearn.linear_model import LinearRegression

# LINEAR REGRESSION ERSTELLEN
lr = LinearRegression()

# MODELL TRAINIEREN
lr.fit(X_train, Y_train)

# PREDICTIONS MACHEN
Y_pred_lr = lr.predict(X_test)
```

```python
# IMPORT LIBRARIES
from sklearn.neighbors import KNeighborsRegressor

# KNEIGHBORS REGRESSOR ERSTELLEN
nr = KNeighborsRegressor()

# MODELL TRAINIEREN
nr.fit(X_train, Y_train)

# PREDICTIONS MACHEN
Y_pred_nr = nr.predict(X_test)
```

### ‚öñÔ∏è Vorhersagen vergleichen

```python
# VERGLEICH DER VORHERGESAGTEN WERTE
comparison_df = pd.DataFrame ({
    'Original' : Y_test,
    'Linear Regression' : Y_pred_lr,
    'KNeighbors Regressor' : Y_pred_nr
}, index=X_test.index)
comparison_df
```
