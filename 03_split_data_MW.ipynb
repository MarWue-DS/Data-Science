# ‚úÇÔ∏è Split Data
### Insurance Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.model_selection import train_test_split

# LOAD DATA
df = pd.read_csv("03_insurance_data.csv")
df
```

```python
# FEATURES & TARGET
X = df.drop(['charges'], axis=1)
Y = df['charges']
```

### ‚úÇÔ∏è Insurance Daten aufteilen
Daten splitten in 70% Training und 30% Testdaten.

```python
# DATEN IN TRAIN UND TEST SPLITTEN
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state=1234)
```

### ‚úÖ Aufgabe Visualisierung
üìà Durchschnittliche Kosten pro Region einmal f√ºr Train und einmal f√ºr Test Daten.

```python
# IMPORT LIBRARIES
import seaborn as sns
import matplotlib.pyplot as plt

# LABELS JE DATASET VERGEBEN
train_df = X_train.copy()
train_df['charges'] = Y_train
train_df['Split'] = 'Train'

test_df = X_test.copy()
test_df['charges'] = Y_test
test_df['Split'] = 'Test'

# COMBINE
combined = pd.concat([train_df, test_df])

# BOXPLOT
plt.figure(figsize=(10, 6))
sns.boxplot(data=combined, x='region', y='charges', hue='Split')
plt.title('Vergleich Kostenverteilung je Region', loc='left', pad=35)
plt.legend(loc='upper left', bbox_to_anchor=(0, 1.09), ncol=3)
plt.show()
```

### Titanic Daten laden

```python
# LOAD DATA
import pandas as pd
from sklearn.model_selection import train_test_split

# LOAD DATA
df = pd.read_csv("03_titanic_data.csv")
df
```

```python
# FEATURES & TARGET
X = df.drop(['Survived'], axis=1)
Y = df['Survived']
```

### ‚úÇÔ∏è Titanic Daten aufteilen
Daten splitten in 70% Training und 30% Testdaten.

```python
# DATEN IN TRAIN UND TEST SPLITTEN
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, stratify=Y, random_state=1234)
```

### ‚úÖ Aufgabe Visualisierung
üìà √úberpr√ºfen, ob die Verteilung der √úberlebenden in den Trainings- und Testsets erhalten geblieben ist

```python
# TRAIN VS. TEST
import seaborn as sns
import matplotlib.pyplot as plt

# TRAIN
train_df = X_train.copy()
train_df['Survived'] = Y_train
train_df['Split'] = 'Train'

# TEST
test_df = X_test.copy()
test_df['Survived'] = Y_test
test_df['Split'] = 'Test'

combined = pd.concat([train_df, test_df])

plt.figure(figsize=(10, 6))
sns.boxplot(data=combined, x='Sex', y='Survived', hue='Split')
plt.title('Vergleich √úberlebende je Geschlecht', loc='left', pad=35)
plt.legend(loc='upper left', bbox_to_anchor=(0, 1.09), ncol=3)
plt.show()
```
