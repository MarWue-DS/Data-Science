# ‚òëÔ∏è Feature Selection

### Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import math
from sklearn.feature_selection import f_regression as fr

# LOAD DATA
df = pd.read_csv('05_students.csv')
df
```

```python
# FEATURES & TARGET
x = df.drop(['Marks'], axis=1)
y = df['Marks']
```

```python
# FEATURES & TARGET NACH F_REGRESSION
#x = df.drop(['Marks'], axis=1)
x = df.drop(['Marks', 'hoursplayed', 'income', 'distance', 'calories'], axis=1)
y = df['Marks']
```

### üí™ Training mit Linearer Regression

```python
# INITIALIZE MODEL
lr = LinearRegression()

# SPLIT DATA
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.4, random_state=1234)
lr.fit(X_train, Y_train)

# MAKE PREDICTIONS
Y_predict = lr.predict(X_test)
```

### üßê Evaluate Predictions
Berechne die Kennzahlen **RMSE** und **R¬≤**.

```python
# CALCULATE RMSE
rmse = math.sqrt(mean_squared_error(Y_test, Y_predict))
print(f"RMSE: {rmse:.2f}")

# CALUCALTE R¬≤
r2 = lr.score(X_test, Y_test)
print(f"R¬≤: {r2:.2f}")
```

### Feature Selection
Feature mittels <code>f_regresion</code> ausw√§hlen.

```python
# RESULT IST LISTE MIT JE 1 ARRAY MIT F-SCORE UND P-VALUE 
result = fr(x, y)
result
```

```python
# CONVERT TO DATAFRAME: SPALTE F-SCORE und P-VALUE
df = pd.DataFrame({'f_score': result[0], 'p_value': result[1]})
df['f_score'] = df['f_score'].map('{:4.2f}'.format)
df['p_value'] = df['p_value'].map('{:2.6f}'.format)
df
```

```python
# SPALTENNAMEN HINZUF√úGEN
columns = list(x.columns)
df['feature'] = columns
df
```
