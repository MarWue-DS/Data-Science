# 03 Daten Transformatieren
Lege eine neue Spalte **DelayCategory** üÜï auf Basis der Versp√§tungen (ArrDelay) an.
Lege eine neue Spalte **Wochentag** üÜï auf Basis der Versp√§tungen (DayOfWeek) an.

```python
# NAMEN F√úR KATEGORIEN VERGEBEN

def categorize_delay(delay):
    if delay < 15:
        return "On Time"
    elif 15 <= delay < 60:
        return "Delay"
    elif 60 <= delay < 360:
        return "Long Delay"
    else:
        return "Extreme Delay"


df_delete['delay_category'] = df_delete['ArrDelay'].apply(categorize_delay)

def categorize_dow(dow):
    if dow < 6:
        return "Weekday"
    else:
        return "Weekend"
        
df_delete['Wochentag'] = df_delete['DayOfWeek'].apply(categorize_dow)        

df_delete
```

Aufteilen der Daten ‚úÇÔ∏è in ein Trainings- und Testdatenset.

```python
# SPLIT in 70% TRAIN und 30% TEST
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df_delete, test_size=0.3, random_state=123)
```
