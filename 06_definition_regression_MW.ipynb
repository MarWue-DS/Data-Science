# ðŸ“ˆ Regression
### Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd
import matplotlib.pyplot as plt

# LOAD DATA
df = pd.read_csv("04_marketing_data.csv")
df
```

### :Â¨Â·. Daten Visualisieren
Visualisere die Daten mit einem Scatterplot - Umsatz vs. Werbungskosten.

```python
# SCATTERPLOT
df.plot.scatter(x="Werbungskosten", y="Umsatz")
```

### ðŸ’ª Modell Trainieren
Trainiere ein lineares Regressionsmodell.

```python
# IMPORT LIBRARIES
from sklearn.linear_model import LinearRegression

# FEATURES & TARGET
x = df[["Werbungskosten"]]
y = df ["Umsatz"]

# TRAIN MODEL
lr = LinearRegression()
lr.fit (x,y)
```

### ðŸ“‰ Gerade zeichnen
Passen die Vorhersagen zu den Datenpunkten?

```python
# PREDICTIONS
Y_pred = lr.predict(x)
```

```python
# SCATTERPLOT
plt.scatter (x, y, color="blue", label="Datenpunkte", edgecolor="none")

# VORGERSAGE GERADE
plt.plot(x, Y_pred, color="red", label="Regressionsgerade")

# LABELS, TITLE, LEGEND
plt.xlabel("Werbungskosten")
plt.ylabel("Umsatz")
plt.title("Lineare Regression")
plt.legend()
# SAVE & SHOW
plt.tight_layout()
plt.savefig("gerade_prediction.png", dpi=300)
plt.show()
```

### â†—ï¸ Steigung und Achsenabschnitt ausgeben

```python
# COEFFICIENT & INTERCEPT
print("Steigung", lr.coef_[0])
print("Achsenabschnitt:", lr.intercept_)
```

```python
# UMSATZ VORHERSAGE
vorhersage = lr.predict(pd.DataFrame({"Werbungskosten": [4000]}))[0]
print(f"Vorhersage: {vorhersage:.2f}â‚¬")
```

### âœï¸ Beurteile dein Modell
Berechnen alle vier Kennzahlen fÃ¼r deine Vorhersagen.

```python
# IMPORT LIBRARIES
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# CALCULATE METRICS
print("MAE:", mean_absolute_error(y, Y_pred))
print("RMSE:", mean_squared_error(y, Y_pred))
print("RAE:", sum(abs(y - Y_pred)) / sum(abs(y - y.mean())))
print("RÂ²:", r2_score(y, Y_pred))
```
