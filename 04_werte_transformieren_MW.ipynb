# üîÑ Daten transformieren
### Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd

# LOAD DATA
df = pd.read_csv("02_wine_data.csv")
df
```

### üîÑ Daten transformieren
Teile den üç∑ Wein in eine Kategorie ein, basierend auf der Spalte quality:
- **kleiner 5:** low
- **kleiner 7:** medium
- **gr√∂√üer 7:** high

```python
# FUNKTION, UM KATEGORIEN ZU VERGEBEN
def categorize (value):
    if value < 5:
        return 'low'
    elif value < 7:
        return 'medium'
    else:
        return 'high'

# FUNKTION ANWENDEN, UM NEUE SPALTE QUALITY_CATEGORY ANZULEGEN
df['quality_category'] = df['quality'].apply(categorize)
df
```

```python
category_counts = df['quality_category'].value_counts()

import matplotlib.pyplot as plt

category_counts.plot(kind="bar", color='navy')
plt.show()
```
