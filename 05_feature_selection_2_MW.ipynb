### Create Data

```python
# IMPORT LIBRARIES
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# CREATE SYNTHETIC DATA
# CREATE SYNTHETIC DATA
df = pd.DataFrame({
    'Fatalities': ['Eins', 'Zwei', 'Mehrere'],
    'Start': [19, 11, 4],
    'Flug': [39, 21, 11],
    'Landung': [19, 6, 8]
})

df
```

```python
# CALCLUATE CHI2
chi2, p, dof, expected = chi2_contingency(df.set_index('Fatalities'))
```

```python
print(f"Chi2: {chi2}\nDOF: {dof}")
print(f"p-Wert: {p}")
```
