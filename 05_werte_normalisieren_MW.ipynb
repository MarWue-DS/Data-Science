# ⚖️ Werte normalisieren
### Daten laden

```python
# IMPORT LIBRARIES
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, RobustScaler
import numpy as np
import matplotlib.pyplot as plt

# LOAD DATA
df = pd.read_csv("03_housing_data.csv")
df
```

### MinMaxScaler
Daten mit <code>MinMaxScaler</code> normalisieren.

```python
# DATEN NORMALISIEREN
minmax = MinMaxScaler().fit_transform(df)
```

### RobustScaler
Daten mit <code>RobustScaler</code> normalisieren.

```python
# DATEN NORMALISIEREN
robust = RobustScaler().fit_transform(df)
```

### Daten Visualisieren
Veranschaulichen, ob Daten der Spalte <code>MedInc</code> **stabil** bleiben oder **verzerrt** werden.

```python
# NUMPY ARRAY IN DATAFRAME MIT SELBEN SPALTENNAMEN UMWANDELN
#df_minmax = pd.DataFrame(minmax, columns=df.columns)
#df_robust = pd.DataFrame(robust, columns=df.columns)

df_minmax = pd.Series(minmax.flatten(), name="MinMaxScaler")
df_robust = pd.Series(robust.flatten(), name="RobustScaler")
df_minmax
```

```python
df_robust
```

```python
# HISTOGRAM MINMAXSCALER
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(df_minmax, bins=30, color='skyblue', alpha=0.7)
plt.title("MinMaxScaler Verteilung")
plt.xlabel("Wertebereich")
plt.ylabel("Anzahl der Einträge")
plt.legend(['MinMaxScaler'])

# HISTOGRAM ROBUSTSCALER
plt.subplot(1, 2, 2)
plt.hist(df_robust, bins=30, color='salmon', alpha=0.7)
plt.title("RobustScaler Verteilung")
plt.xlabel("Wertebereich")
plt.ylabel("Anzahl der Einträge")
plt.legend(['RobustScaler'])

# LEGENDE
plt.tight_layout()

# SHOW PLOT
plt.show()
```
